```python
df = pd.read_csv('survey lung cancer.csv', delimiter=',', header=0, encoding='utf-8')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder = LabelEncoder()
```

```python
for label_name in df.columns.tolist()[2:]:
    df[label_name] = LabelEncoder.fit_transform(df[label_name])
```

```python
for label_name in df.columns.tolist()[0:1]:
    df[label_name] = LabelEncoder.fit_transform(df[label_name])    
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df.pop('LUNG_CANCER')
X = df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(X)
```

```python

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
fn = LogisticRegression()
```

```python
fn.fit(X_train,y_train)
```

```python
predict = fn.predict(X_test)
```

```python
report = classification_report(y_test, predict)
```

```python
accuracy = round(accuracy_score(y_test, predict) * 100, ndigits = 2)
```
